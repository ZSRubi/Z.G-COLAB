```python
#ejercicio2
class Producto:
    def __init__(self, nombre, precio, cantidad):
        self.nombre = nombre
        self.precio = precio
        self.cantidad = cantidad

    def calcular_valor(self):
        return self.precio * self.cantidad

class ProductoPerecedero(Producto):
    def calcular_valor(self):
        total = super().calcular_valor()
        if self.cantidad > 10:
            total *= 0.9
        return total

class ProductoNoPerecedero(Producto):
    def calcular_valor(self):
        total = super().calcular_valor()
        if self.cantidad > 10:
            total *= 0.95
        return total

inventario = [
    ProductoPerecedero("Leche", 3.5, 12),
    ProductoNoPerecedero("Lata de atún", 2.0, 8),
    ProductoPerecedero("Pan", 1.0, 15)
]

valor_total = 0
for producto in inventario:
    valor = producto.calcular_valor()
    print(f"{producto.nombre}: ${valor:.2f}")
    valor_total += valor

print(f"\nValor total del inventario: ${valor_total:.2f}")
```

```python
#ejercicio3
class CuentaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.saldo = saldo

    def mostrar_tipo(self):
        return "Cuenta Bancaria"

    def depositar(self, monto):
        if monto > 0:
            self.saldo += monto
            print(f"Depósito exitoso. Nuevo saldo: ${self.saldo}")
        else:
            print("El monto debe ser positivo.")

    def retirar(self, monto):
        if 0 < monto <= self.saldo:
            self.saldo -= monto
            print(f"Retiro exitoso. Nuevo saldo: ${self.saldo}")
        else:
            print("Fondos insuficientes o monto inválido.")

    def consultar_saldo(self):
        print(f"Saldo actual: ${self.saldo}")

class CuentaAhorro(CuentaBancaria):
    def mostrar_tipo(self):
        return "Cuenta de Ahorro"

class CuentaCorriente(CuentaBancaria):
    def mostrar_tipo(self):
        return "Cuenta Corriente"

cuenta = CuentaAhorro("Juan", 1000)


while True:
    print("\n--- Menú ---")
    print("1. Consultar saldo")
    print("2. Depositar")
    print("3. Retirar")
    print("4. Mostrar tipo de cuenta")
    print("5. Salir")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        cuenta.consultar_saldo()
    elif opcion == "2":
        monto = float(input("Ingrese monto a depositar: "))
        cuenta.depositar(monto)
    elif opcion == "3":
        monto = float(input("Ingrese monto a retirar: "))
        cuenta.retirar(monto)
    elif opcion == "4":
        print(cuenta.mostrar_tipo())
    elif opcion == "5":
        print("Saliendo del sistema...")
        break
    else:
        print("Opción inválida.")
```

```python
#ejercicio4
class Empleado:
    def __init__(self, nombre, salario_base):
        self.nombre = nombre
        self.salario_base = salario_base

    def calcular_salario(self):
        return self.salario_base

class EmpleadoTiempoCompleto(Empleado):
    def calcular_salario(self):
        bono = 500 if self.salario_base > 2000 else 300
        return self.salario_base + bono

class EmpleadoMedioTiempo(Empleado):
    def calcular_salario(self):
        bono = 200 if self.salario_base > 1000 else 100
        return self.salario_base + bono

empleados = [
    EmpleadoTiempoCompleto("Carlos", 2200),
    EmpleadoMedioTiempo("Ana", 900),
    EmpleadoTiempoCompleto("Luis", 1800),
    EmpleadoMedioTiempo("Marta", 1100)
]


for empleado in empleados:
    salario_total = empleado.calcular_salario()
    print(f"{empleado.nombre} - Salario total: ${salario_total}")
```

```python
#ejercicio5
import math

class Figura:
    def area(self):
        return 0

class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura

class Triangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return (self.base * self.altura) / 2

class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * self.radio ** 2


figuras = [
    Rectangulo(5, 3),
    Triangulo(4, 6),
    Circulo(2)
]


for figura in figuras:
    print(f"Área: {figura.area():.2f}")
```

```python
#ejercicio6
class Mascota:
    def __init__(self, nombre):
        self.nombre = nombre

    def hacer_sonido(self):
        return "Sonido genérico"

class Perro(Mascota):
    def hacer_sonido(self):
        return "Guau guau"

class Gato(Mascota):
    def hacer_sonido(self):
        return "Miau"

class Pez(Mascota):
    def hacer_sonido(self):
        return "..."

mascotas = []


while True:
    print("\n--- Menú de mascotas ---")
    print("1. Agregar mascota")
    print("2. Listar mascotas")
    print("3. Salir")

    opcion = input("Elige una opción: ")

    if opcion == "1":
        nombre = input("Nombre de la mascota: ")
        tipo = input("Tipo (perro/gato/pez): ").lower()

        if tipo == "perro":
            mascotas.append(Perro(nombre))
        elif tipo == "gato":
            mascotas.append(Gato(nombre))
        elif tipo == "pez":
            mascotas.append(Pez(nombre))
        else:
            print("Tipo no reconocido.")
    elif opcion == "2":
        print("\nListado de mascotas:")
        for mascota in mascotas:
            print(f"{mascota.nombre} dice: {mascota.hacer_sonido()}")
    elif opcion == "3":
        print("¡Hasta luego!")
        break
    else:
        print("Opción inválida.")
```

```python
#ejercicio7
class Estudiante:
    def _init_(self, nombre, notas):
        self.nombre = nombre
        self.notas = notas

    def calcular_promedio(self):
        suma = 0
        for nota in self.notas:
            suma += nota
        return suma / len(self.notas)

    def esta_aprobado(self):
        promedio = self.calcular_promedio()
        if promedio >= 11:
            return True
        else:
            return False

    def mostrar_resultado(self):
        print(f"{self.nombre} tiene un promedio de {self.calcular_promedio():.2f}")
        if self.esta_aprobado():
            print("Estado: Aprobado")
        else:
            print("Estado: Desaprobado")



class EstudianteBecado(Estudiante):
    def esta_aprobado(self):
        promedio = self.calcular_promedio()
        if promedio >= 14:
            return True
        else:
            return False



e1 = Estudiante("Carlos", [12, 10, 13])
e2 = EstudianteBecado("María", [13, 14, 12])

e1.mostrar_resultado()
print()
e2.mostrar_resultado()
```

```python
#ejercicio8
class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

    def __str__(self):
        return f"{self.nombre} - ${self.precio}"

class Cliente:
    def __init__(self, nombre):
        self.nombre = nombre
        self.carrito = []

    def agregar_producto(self, producto):
        self.carrito.append(producto)

    def calcular_total(self):
        total = sum(p.precio for p in self.carrito)
        if len(self.carrito) > 3:
            total *= 0.9
        return total

    def mostrar_carrito(self):
        for producto in self.carrito:
            print(producto)


productos = [
    Producto("Laptop", 1000),
    Producto("Smartphone", 800),
    Producto("Auriculares", 150),
    Producto("Teclado", 120),
    Producto("Ratón", 50)
]

cliente = Cliente("Juan")
while True:
    print("\nProductos disponibles:")
    for i, producto in enumerate(productos):
        print(f"{i+1}. {producto}")
    opcion = int(input("\nSelecciona el producto a comprar (0 para salir): "))
    if opcion == 0:
        break
    cliente.agregar_producto(productos[opcion-1])
    print(f"Producto agregado: {productos[opcion-1].nombre}")

    continuar = input("¿Quieres seguir comprando? (s/n): ")
    if continuar.lower() != 's':
        break

print("\nCarrito de compras:")
cliente.mostrar_carrito()
print(f"Total a pagar: ${cliente.calcular_total():.2f}")
```

```python
#ejercicio11
import math

class Figura:
    def area(self):
        pass

    def perimetro(self):
        pass

class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura

    def perimetro(self):
        return 2 * (self.base + self.altura)

class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * (self.radio ** 2)

    def perimetro(self):
        return 2 * math.pi * self.radio


figuras = [
    Rectangulo(10, 5),
    Circulo(7),
    Rectangulo(4, 3),
    Circulo(2)
]


for figura in figuras:
    print(f"Área: {figura.area():.2f}, Perímetro: {figura.perimetro():.2f}")
```

```python
#ejercicio14
class Producto:
    def __init__(self, nombre, precio, stock):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock

    def descuento(self):
        pass

    def evaluar_stock(self):
        if self.stock < 5:
            print(f"El producto {self.nombre} está por agotarse.")
        else:
            print(f"Stock suficiente de {self.nombre}.")

class Ropa(Producto):
    def descuento(self):
        return self.precio * 0.8

class Alimento(Producto):
    def descuento(self):
        return self.precio * 0.9

class Tecnologia(Producto):
    def descuento(self):
        return self.precio * 0.7


productos = [
    Ropa("Camiseta", 20, 3),
    Alimento("Arroz", 10, 15),
    Tecnologia("Laptop", 1000, 2),
    Ropa("Pantalón", 25, 8)
]


for producto in productos:
    print(f"{producto.nombre} - Precio con descuento: ${producto.descuento():.2f}")
    producto.evaluar_stock()
```

```python
#ejercicio15
class Persona:
    def rol(self):
        pass

class Estudiante(Persona):
    def __init__(self, nombre, notas):
        self.nombre = nombre
        self.notas = notas

    def rol(self):
        return "Estudia"

    def promedio(self):
        return sum(self.notas) / len(self.notas)

class Profesor(Persona):
    def rol(self):
        return "Enseña"


estudiantes = [
    Estudiante("Juan", [8, 9, 10]),
    Estudiante("Ana", [5, 6, 7]),
    Estudiante("Luis", [9, 9, 8])
]

profesores = [
    Profesor(),
    Profesor()
]


for estudiante in estudiantes:
    print(f"{estudiante.nombre} - Promedio: {estudiante.promedio():.2f}")
    if estudiante.promedio() >= 6:
        print(f"{estudiante.nombre} ha aprobado.")
    else:
        print(f"{estudiante.nombre} no ha aprobado.")


for profesor in profesores:
    print(f"Rol del profesor: {profesor.rol()}")
```
